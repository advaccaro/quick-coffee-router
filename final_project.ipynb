```python
# Import libraries
import pandas as pd
import numpy as np
from datetime import datetime
import googlemaps #Geocoding and travel times
import gmaps #Visualization and routing
import populartimes #Popularity scores, 
```

```python
# Set up Google API
google_api_key = ''
gmc = googlemaps.Client(key=google_api_key)
gmaps.configure(api_key=google_api_key)
```

```python
# Path to Starbucks csv file
data = './data/Starbucks_US_Addresses_2013.csv'
```

```python
# Read in CSV
raw = pd.read_csv(data)
```

```python
# Clean and filter data
reduced = raw[['Street Line 1', 'Street Line 2','City', 'State']]
pasadena = reduced.loc[(reduced['City'] == 'Pasadena') & (reduced['State'] == 'CA')]
```

```python
# Add complete address to dataframe
addresses = []
for index, row in pasadena.iterrows():
    if str(row[1]) == 'nan':
        addresses.append(', '.join(['Starbucks ' + row[0], row[2], row[3]]))
    else:
        addresses.append(', '.join(['Starbucks ' + row[0], str(row[1]), row[2], row[3]]))
pasadena.loc[:, 'Address'] = pd.Series(addresses, index=pasadena.index)
p = pasadena.drop(['Street Line 1', 'Street Line 2', 'City', 'State'], axis=1)
p
```

```python
### Estimate time to get coffee
### and classify locations according to wait time/open/closed
time_now = datetime.now()
day_of_week = datetime.weekday(time_now)
hour_now = time_now.hour
#hour_now = 8
day_of_week, hour_now
row = p.iloc[0,:]
row['popular_times'][day_of_week]['data'][hour_now]
locations = []
waits = []
good_wait = []
ok_wait = []
bad_wait = []
very_bad_wait = []
closed = []
for index, row in p.iterrows():
    loc = (row['Latitude'], row['Longitude'])
    locations.append(loc)
    popularity = row['popular_times'][day_of_week]['data'][hour_now]
    avg_time_spent = row['time_spent'][0]
    if popularity == 0:
        wait = 0
    else:
        wait = (popularity/100 + .5) * avg_time_spent
    waits.append(wait)
    if wait == 0:
        closed.append(loc)
    elif wait > 0 and wait <= 8:
        good_wait.append(loc)
    elif wait > 8 and wait <= 12:
        ok_wait.append(loc)
    elif wait > 12 and wait <= 16:
        bad_wait.append(loc)
    elif wait > 16:
        very_bad_wait.append(loc)
        
```

```python
### Add estimated time to get coffee to DataFrame
p.loc[:, 'Coffee time'] = pd.Series(waits, index=p.index)
p
```

```python
### Remove rows that are closed or do not have coffee time available
p = p[p['Coffee time'] > 0]
p
```

```python
### Test driving time retrieval
dist = gmc.distance_matrix(start_location, end_location, departure_time='now')
dist
```

```python
### Extract travel time from distance_matrix results
seconds_total = dist['rows'][0]['elements'][0]['duration_in_traffic']['value']
minutes_est = dist['rows'][0]['elements'][0]['duration_in_traffic']['text']
print(minutes_est, seconds_total)
```

```python
### Define function to run distance_matrix and extract results
def time_from_loc1_to_loc2(loc1, loc2):
    dist = gmc.distance_matrix(loc1, loc2, departure_time='now')
    seconds_total = dist['rows'][0]['elements'][0]['duration_in_traffic']['value']
    minutes_est = np.round(seconds_total/60)
    return(minutes_est, seconds_total)
```

```python
### Test new function
minutes_est, seconds_total = time_from_loc1_to_loc2(start_location, end_location)
print(minutes_est, seconds_total)
```

```python
### Define function to calculate total travel time
def time_from_loc1_to_loc2_to_loc3(loc1, loc2, loc3=False):
    if not loc3:
        loc3=loc1
    journey1, seconds_total1 = time_from_loc1_to_loc2(loc1, loc2)
    journey2, seconds_total2 = time_from_loc1_to_loc2(loc2, loc3)
    total_journey = np.round((seconds_total1 + seconds_total2)/60, decimals=1)
    return(total_journey)
```

```python
### Calculate travel time for all locations
p['Travel time'] = p['Address'].apply(lambda x: time_from_loc1_to_loc2_to_loc3(start_location, x, end_location))
```
